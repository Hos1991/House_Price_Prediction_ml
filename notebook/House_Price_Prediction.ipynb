# Modeling

## Defining Proper Features - Linear Regression

```python
features = [
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "TotalBsmtSF",
    "Age",
    "LotArea_log"
]
```

```python
# The above features are appropriate for linear regression model 
```

```python
x = df[features]
y = df["SalePrice"]
```

```python
x.isnull().sum()
```

## Linear Regression

### Spliting Train and Test Datasets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Using Linear Regression to model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

## Evaluation

### Using Mean Absolute Error - Linear Regression

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

## Defining Proper Features - Random Forest

```python
features = [
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "TotalBsmtSF",
    "Age",
    "LotArea_log",
    "TotRmsAbvGrd",
    "FullBath",
    "GarageArea"
]
```

```python
# Three more features are added for Random Forest model 
```

```python
x = df[features]
y = df["SalePrice"]
```

## Random Forest

### Spliting Train and Test Datasets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Using Random Forest to model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state = 42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

## Evaluation

### Using Mean Absolute Error - Random Forest

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

# Feature Importance

```python
importance = model.feature_importances_
feature_importance = pd.Series(importance, index=features)
print(feature_importance.sort_values(ascending = False))
```
